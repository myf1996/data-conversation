```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('sample_file/books.xml')
root = tree.getroot()
```

```python
root
```

```python
for elem in root:
    print(elem.tag,".....",elem.text)
    for subelem in elem:
        print(subelem.text,".....",subelem.tag)
        for i in subelem:
            print(i.text,",,,,,,,,,,,,,,,,,,,,,,,,,,,")
```

```python
child = root.getchildren()

for i in child:
    print(i,"---------------------------------ROOT")
    for j in i.getchildren():
        print(j,"----------------------------CHILD")
        for k in j.getchildren():
            print(k,"---------------------SUB-CHILD")
```
